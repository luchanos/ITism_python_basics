
# Условные конструкции в Python

Условные конструкции позволяют вашей программе выполнять различные действия в зависимости от условий. Python поддерживает обычные условные конструкции, включая `if`, `elif`, и `else`.



## Теоретический обзор

- **`if`** используется для проверки условия и выполнения блока кода, если условие истинно.
- **`elif`** (сокращение от else if) проверяет другое условие, если предыдущие условия ложны.
- **`else`** выполняет блок кода, если ни одно из предыдущих условий не было истинно.

Синтаксис условных конструкций в Python:

```python
if условие1:
    # выполняется, если условие1 истинно
elif условие2:
    # выполняется, если условие1 ложно, но условие2 истинно
else:
    # выполняется, если условие1 и условие2 ложны
```



## Примеры кода

Рассмотрим несколько простых примеров использования условных конструкций.


````python

age = 18

if age >= 18:
    print("Вы взрослый!")
elif age > 13:
    print("Вы подросток!")
else:
    print("Вы ребенок!")
````
